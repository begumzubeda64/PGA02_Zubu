```python
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import numpy as np

df_insurance = pd.read_csv(r"C:\zubeda\PGA-02\Python\Datasets\Visualization\insurance_data_with_day.csv")

frequency_by_region = df_insurance["region"].value_counts()
keys = frequency_by_region.keys().to_list()
counts = frequency_by_region.to_list()
cs = cm.Set1([2, 4, 6, 8])

plt.pie(x=counts, labels=keys, autopct='%1.1f%%', colors=cs)
plt.title("Region-wise Distribution")
plt.show()
```

```python
frequency_by_region = df_insurance["region"].value_counts()
keys = frequency_by_region.keys().to_list()
counts = frequency_by_region.to_list()
cs = cm.Set1([1, 5, 3, 7])
explode = [0.1, 0, 0, 0]

plt.pie(x=counts, labels=keys, autopct='%1.1f%%', colors=cs, explode=explode)
plt.title("Region-wise Distribution")
plt.show()
```

```python
frequency_by_region = df_insurance["region"].value_counts()
keys = frequency_by_region.keys().to_list()
counts = frequency_by_region.to_list()

plt.pie(x=counts, labels=keys, autopct='%1.1f%%')
circle = plt.Circle(xy=(0, 0), radius=0.4, color="white")
plt.gcf()
plt.gca().add_artist(circle)
plt.title("Region-wise Distribution")
plt.show()
```

```python
plt.scatter(x="bmi", y="claim", data=df_insurance)

plt.xlabel("BMI")
plt.ylabel("Claim")
plt.title("Relationship between BMI & Claim")
plt.show()
```

```python
import seaborn as sns

df_titanic = pd.read_csv(r"C:\zubeda\PGA-02\Python\Datasets\Visualization\Titanic_data.csv")
df_titanic.head()
```

```python
sns.stripplot(x="Sex", y="Age", data=df_titanic)

plt.title("Strip plot for Gender and Age")
plt.show()
```

```python
sns.stripplot(x="Sex", y="Age", hue="Survived", data=df_titanic)

plt.title("Survival based on Age and Gender")
plt.show()
```

```python
sns.distplot(a=df_insurance["claim"], color="olive")

plt.title("Distribution of Insurance Claim")
plt.show()
```

```python
sns.distplot(a=df_insurance["claim"], color="m", kde=False)

plt.title("Distribution of Insurance Claim")
plt.show()
```

```python
sns.jointplot(x="bmi", y="claim", data=df_insurance, color="sandybrown")

plt.show()
```

```python
sns.violinplot(x="Sex", y="Age", data=df_titanic)

plt.title("Violin plot for Age and Gender")
plt.show()
```

```python
sns.violinplot(x="Sex", y="Age", hue="Survived", data=df_titanic, split=True)

plt.title("Survival based on Age and Gender")
plt.show()
```

```python
sns.swarmplot(x="Sex", y="Age", data=df_titanic)

plt.title("Swarm plot for Age and Gender")
plt.show()
```

```python
sns.swarmplot(x="Sex", y="Age", hue="Survived", data=df_titanic)

plt.title("Survival based on Age and Gender")
plt.show()
```
