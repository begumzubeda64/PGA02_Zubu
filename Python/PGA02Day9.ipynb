```python
import numpy as np

month = np.arange(1, 13)
prices = [101, 112, 203, 304, 335, 406, 507, 608, 709, 810, 795, 854]

plt.plot(month, prices, color='b', marker='*')
plt.xlabel('Month')
plt.ylabel('Prices')
plt.title("Price Trend in a Year")
#plt.grid() #grid based background
plt.grid(linestyle='--', linewidth='0.5', color='g')

plt.show()
```

```python
import pandas as pd

df_insurance = pd.read_csv(r"C:\zubeda\PGA-02\Python\Datasets\Visualization\insurance_data_with_day.csv")
df_ins_groupby_day = df_insurance.groupby("dayofmonth")
days = df_ins_groupby_day["dayofmonth"].unique()

plt.plot(days, df_ins_groupby_day['age'].mean(), color='limegreen', marker='*', label='Age')
plt.plot(days, df_ins_groupby_day['bmi'].mean(), color='crimson', marker='*', label='BMI')
plt.plot(days, df_ins_groupby_day['bloodpressure'].mean(), color='goldenrod', marker='*', label='Blood Pressure')

plt.title("Multi-line Plots")
plt.xlabel("Days")
plt.ylabel("Measure")
plt.legend(loc="center right")
plt.show()
```

```python
#Days wise average claim of gender
fig, ax = plt.subplots(figsize=(7, 4)) #height, width

plt.style.use("ggplot")
df_ins_groupby_region_gender = df_insurance.groupby(["dayofmonth", "gender"])
df_ins_groupby_region_gender.mean()["claim"].unstack().plot(ax=ax) #unstack makes groupby gender as column

plt.show()
```

```python
fig, ax = plt.subplots(2,2) #row,col
x = np.linspace(0, 8, 1000)

ax[0, 0].plot(x, x**2, color='g')
ax[0, 0].title.set_text("Axis [0, 0]")

ax[1, 0].plot(x, x**3, color='k')
ax[1, 0].title.set_text("Axis [1, 0]")

ax[0, 1].plot(x, np.sin(x), color='b')
ax[0, 1].title.set_text("Axis [0, 1]")

ax[1, 1].plot(x, np.cos(x), color='r')
ax[1, 1].title.set_text("Axis [1, 1]")

fig.show()
```

```python
#grouped bar plot side by side
Python_marks = (50, 65, 40, 35, 77)
R_marks = (55, 72, 94, 70, 85)

index = np.arange(5)

plt.bar(x=index, height=Python_marks, width=0.35, label="Python")
plt.bar(x=index + 0.35, height=R_marks, width=0.35, label="R")

plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Scores by Students")
plt.legend()
plt.show()
```

```python
#stacked bar plot
Python_marks = (50, 65, 40, 35, 77)
R_marks = (55, 72, 94, 70, 85)

index = np.arange(5)

plt.bar(x=index, height=Python_marks, width=0.35, label="Python")
plt.bar(x=index, height=R_marks, width=0.35, bottom=Python_marks, label="R")

plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Scores by Students")
plt.legend()
plt.xticks(ticks=index, labels=("A", "B", "C", "D", "E"))
plt.show()
```
