```python
score_data = {"Name": ["James", "Mia", "Emity", "Roben", "John", "Jordan", "Bima"], 
              "Verbal_Score": [151, 152, 151, 156, 100, 145, 155], 
              "Quantitative_Score": [158, 87, 100, 147, 139, 129, 122], 
              "Qualify": ["Yes", "No", "No", "Yes", "No", "Yes", "Yes"]}

df_score = pd.DataFrame(score_data)
print(df_score)
```

```python
#rank() - ranks values calculate average and assign same values for same rank
df_score["Verbal_Rank"] = df_score.Verbal_Score.rank()
df_score
```

```python
#rank() - ranks values selects minimum value for same rank and skip the max rank
df_score["Verbal_Rank"] = df_score.Verbal_Score.rank(method="min")
df_score
```

```python
#rank() - ranks values selects maximum value for same rank and skip the min rank
df_score["Verbal_Rank"] = df_score.Verbal_Score.rank(method="max")
df_score
```

```python
#rank() - ranks values assigns same value for same rank and does not skip any rank
df_score["Verbal_Rank"] = df_score.Verbal_Score.rank(method="dense")
df_score
```

```python
print(df_score.sort_values("Verbal_Rank"))
```
