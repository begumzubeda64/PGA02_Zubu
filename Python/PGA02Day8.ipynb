```python
data = { "Cars": ["BMW", "Ford", "Honda", "Volvo", "BMW", "Ford", "BMW"],
       "Sales": [50000, 45000, 51000, 60000, 41000, 21450, 34000],
       "Color": ["Black", "Blue", "Blue", "Black", "Blue", "Black", "Blue"] }

df_cars = pd.DataFrame(data)
df_cars
```

```python
df_insurance = pd.read_csv(r"C:\zubeda\PGA-02\Python\Datasets\Dataframe Operations\insurance_data_with_day.csv")
print(df_insurance.shape)

df_insurance.drop(index=range(2), inplace=True) #permanent delete
print(df_insurance.shape)
```

```python
df_insurance.drop(index=[2, 4], inplace=True)
df_insurance.head()
```

```python
df_insurance.drop('patientid', axis=1, inplace=True) #drop column
df_insurance.head()
```

```python
df_insurance = df_insurance.replace(to_replace="northwest", value="North West")
df_insurance.head()
```

```python
df_insurance = df_insurance.replace(to_replace=["northeast", "southeast", "southwest"], 
                                    value=["North East", "South East", "South West"])
df_insurance.tail(5)
```

```python
#replace by col name
df_insurance["smoker"] = df_insurance["smoker"].replace(to_replace=["Yes", "No"], value=["Smoker", "Non Smoker"])
df_insurance.tail(6)
```

```python
df_insurance.insert(4, 'high bmi', np.nan)
df_insurance.head()
```

```python
#replace by condition
df_insurance.loc[df_insurance['bmi'] > 32, 'high bmi'] = 'Yes'
df_insurance.loc[df_insurance['bmi'] <= 32, 'high bmi'] = 'No'
df_insurance
```

```python
df_insurance["high bmi"] = np.where((df_insurance['bmi'] > 32), 'Yes', df_insurance['high bmi'])
df_insurance["high bmi"] = np.where((df_insurance['bmi'] <= 32), 'No', df_insurance['high bmi'])

df_insurance
```
